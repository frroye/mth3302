## Chargment des librairies

```julia
using CSV, DataFrames, Gadfly, LinearAlgebra, Statistics, Distributions
```

## Chargement des données

```julia
data = CSV.read("fishweights.csv", DataFrame)
first(data,5)
df = filter(row -> (row.Species == "Brème"), data )
first(df, 5)

y = df[:,:TotalLength]
ȳ = mean(y)
s = std(y)
n = length(y)
v = 42.5
α = 1
β = 1
```

## Loi conditionnelle complète de $\mu$.

```julia
f₁(σ²::Real) = Normal((n*ȳ+v)/(n+1),σ²/(n+1))
```

## Loi conditionnelle complète de $\sigma^2$

```julia
f₂(μ::Real) = InverseGamma(α+(n+1)/2, β+ ((μ-v)^2)/2+ sum( (y[i]-μ)^2 for i=1:n)/2 )
```

## Générez un échantillon aléatoire de la loi a posteriori, avec l'échantillonnage de Gibs

```julia
m = 1000

μ = zeros(m)
σ² = zeros(m)

# Initialisation de μ et σ² à des états quelconques

μ[1] = 10
σ²[1] = 10

for i=2:m
    
    μ[i] = rand( f₁(σ²[i-1]) )

    σ²[i] = rand( f₂(μ[i]) )
    
end
```

## Chaîne obtenue pour $\mu$

```julia
plot(x=1:length(μ), y=μ, Geom.line, Guide.xlabel("itération"), Guide.ylabel("μ"))
```

## Chaîne obtenue pour $\sigma^2$

```julia
plot(x=1:length(σ²), y=σ², Geom.line, Guide.xlabel("itération"), Guide.ylabel("σ²"))
```

# Rejeter les itérations de la phase de chauffe

Visuellement, il n'est pas évident d'identifier la phase de chauffe. On retire donc les 100 premières valeurs.

```julia
deleteat!(μ,(1:100))
deleteat!(σ²,(1:100))
```

## Visualiser l'échantillon généré

```julia

plot(
    layer(x=μ, Geom.histogram(bincount=50, density=true)),
    Guide.xlabel("μ"), Guide.ylabel("density")
)
```

```julia

plot(
    layer(x=σ², Geom.histogram(bincount=50, density=true)),
    Guide.xlabel("σ²"), Guide.ylabel("density")
)
```

## Identification de l'intervalle de crédibilité pour  $\mu$

L'échantillon généré semble symétrique. Il n'est donc pas nécessaire d'utiliser l'intervalle de crédibilité HPD.

```julia
alpha = 0.05
I = quantile(μ, [alpha/2, 1-alpha/2])
println("L'intervalle de crédibilité à 95% pour μ est [$(I[1]), $(I[2])].")
```

## Moyenne et Mod de la longueur des brêmes

```julia
moyenneμ = mean(μ)
modeμ = mode(μ)
println("moyenne = $moyenneμ  mode = $modeμ")
```
